```python
opiniones = []  # Crear una lista vacía
contador = 0  # Inicializar el contador

while contador < 50:  # Realizar un máximo de 20 clics en el botón "Siguiente"
    tr_elements = driver.find_elements(By.XPATH, '//*[@id="cm_cr-review_list"]')

    # Iterar sobre los elementos en tr_elements
    for i in range(10):
        opinion_div = tr_elements[0].find_elements(By.XPATH, './div/div/div/div[4]')[i]
        opinion = opinion_div.text
        opiniones.append(opinion)

    # Buscar y hacer clic en el botón "Siguiente"
    try:
        next_button = driver.find_element(By.XPATH, '//*[@id="cm_cr-pagination_bar"]/ul/li[2]/a')
        next_button.click()
        time.sleep(2)  # Esperar a que se carguen las nuevas opiniones (ajusta el tiempo de espera según sea necesario)
        contador += 1  # Incrementar el contador después de hacer clic en el botón
    except:
        break  # Salir del bucle si no hay más páginas de opiniones

# Imprimir la lista de opiniones
print(opiniones)
```

```python
from langdetect import detect
from langdetect.lang_detect_exception import LangDetectException



# Filtrar las opiniones que no están en español
opiniones_espanol = []

for opinion in opiniones:
    try:
        if detect(opinion) == 'es':
            opiniones_espanol.append(opinion)
    except LangDetectException:
        pass

# Imprimir las opiniones en español
for opinion in opiniones_espanol:
    print(opinion)
```

```python
import json

# Cargar las opiniones desde el archivo JSON
with open('opinionesespañol.json', 'r', encoding='utf-8') as archivo:
    datos_json = json.load(archivo)

# Almacenar las opiniones en una lista
opiniones_cargadas = list(datos_json)
```

```python
# Obtener el número de elementos en la lista
num_elementos = len(opiniones_cargadas)
# Imprimir el número de elementos
print("Número opiniones recopiladas:", num_elementos)
```

```python
import re
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

# Expresión regular para eliminar la puntuación y caracteres especiales sin eliminar emojis
pattern = r"[^\w\s\d]"

# Obtener la lista de stopwords en español
stopwords_es = set(stopwords.words('spanish'))

# Lista de opiniones después de eliminar la puntuación y caracteres especiales
opiniones_sin_puntuacion = []

for opinion in opiniones_cargadas:
    # Eliminar la puntuación y caracteres especiales utilizando la expresión regular
    opinion_limpia = re.sub(pattern, "", opinion)
    opiniones_sin_puntuacion.append(opinion_limpia)

# Lista de opiniones convertidas a minúsculas
opiniones_minusculas = []

for opinion in opiniones_sin_puntuacion:
    # Convertir el texto a minúsculas
    opinion_minuscula = opinion.lower()
    opiniones_minusculas.append(opinion_minuscula)

# Lista de opiniones sin stopwords
opiniones_sin_stopwords = []

for opinion in opiniones_minusculas:
    # Dividir el texto en palabras
    palabras = opinion.split()
    # Filtrar las stopwords
    palabras_filtradas = [palabra for palabra in palabras if palabra not in stopwords_es]
    # Unir las palabras filtradas en una sola cadena
    opinion_sin_stopwords = ' '.join(palabras_filtradas)
    opiniones_sin_stopwords.append(opinion_sin_stopwords)

```

```python
print(opiniones_sin_stopwords)
```

```python
# Combinar todas las opiniones en una sola lista de palabras
words = [word.lower() for opinion in opiniones_sin_stopwords for word in opinion.split()]

# Contar la frecuencia de cada palabra
word_count = Counter(words)

# Obtener las 5 palabras más repetidas
top_words = word_count.most_common(10)

# Imprimir los resultados
for word, count in top_words:
    print(word, "-", count)
```

Que entre las palabras mas repetidas este entre ellas las palabras "bien" y "mejor" es algo muy significativo para saber la satisfaccion de los clientes con la compra 

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer



# Descargar el modelo VADER
nltk.download('vader_lexicon')

# Crear una instancia del SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Obtener las palabras negativas de cada opinión
negative_words = []
for opinion in opiniones_sin_stopwords:
    words = opinion.split()
    for word in words:
        scores = analyzer.polarity_scores(word.lower())
        if 'compound' in scores and scores['compound'] < 0:
            negative_words.append(word.lower())

# Contar la frecuencia de cada palabra negativa
negative_word_count = nltk.FreqDist(negative_words)

# Obtener las 5 palabras más negativas
top_negative_words = negative_word_count.most_common(10)

# Imprimir los resultados
for word, count in top_negative_words:
    print(word, "-", count)
```

Hay palabras que detecta como negativas que son de nombres de videojuegos por lo tanto el numero de palabras negativas se reduce mucho mas. La palabra brutal seria algo positivo por lo tanto tambien se descartaria.

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer


# Descargar el modelo VADER
nltk.download('vader_lexicon')

# Crear una instancia del SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Obtener las palabras positivas de cada opinión
positive_words = []
for opinion in opiniones_sin_stopwords:
    words = opinion.split()
    for word in words:
        scores = analyzer.polarity_scores(word.lower())
        if 'compound' in scores and scores['compound'] > 0:
            positive_words.append(word.lower())

# Contar la frecuencia de cada palabra positiva
positive_word_count = nltk.FreqDist(positive_words)

# Obtener las 5 palabras más positivas
top_positive_words = positive_word_count.most_common(10)

# Imprimir los resultados
for word, count in top_positive_words:
    print(word, "-", count)
```

Hay palabras que las detecta como positivas cuando no lo son, aunque basta ver la cantidad de palabras positivas comparando las negativas para saber que la ps5 fue un exito y que los clientes estan satifechos con la compra.
